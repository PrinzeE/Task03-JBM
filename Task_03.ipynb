```python
import paho.mqtt.client as mqtt
from pymongo import MongoClient
client = MongoClient("localhost",27017)
db = client.emphase_neel5
collection = db['test_jbm']
import json

# Define callback function for when a message is received
def on_message(client, userdata, message):
    # Extract data from MQTT message
    topic = message.topic
    payload = message.payload.decode()

    # Process data (replace with your own code)
    print("Received message on topic {}: {}".format(topic, payload))
    print(type(payload))
    collection.insert_one(json.loads(payload))

# Connect to MQTT broker
client = mqtt.Client()
client.connect("3.7.85.13", 1883)

# Subscribe to MQTT topic and start listening for messages
client.subscribe("JBMGroup/em3phase/neel5")
client.on_message = on_message
client.loop_forever()
```
